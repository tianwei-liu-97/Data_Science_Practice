```python
import csv
with open('gapminder.csv') as file:
    gap_data = []
    for row in csv.reader(file):
        gap_data.append(row)
```

```python
def select(data, variable):
    '''
    This is the select() function I wrote.
    
    To use this function, you need to input two things:
    1) data. In this particular case, as we only have the "gap_data" dataset, please enter gap_data. 
    2) variable. In the gap_data dataset, there are three variables: "country", "lifeExp", and "gdpPercap".
    Please use variable = "variable_name" (do not forget quotation marks!) to input the second element to run the function. 
    Note that variable names are case sensitive.
    
    I also try to distinguish between "country" and "lifeExp" & "gdpPercap". 
    Currently, all these elements are stored as strings within lists.
    I try to maintain the string type for "country" because country names are characters.
    I try to convert "lifeExp" and "gdpPercap" into float as they are numbers (not integers also).
    That way, I am hoping the output is readily usable for further analysis.

    '''
    return_list = []
    if variable in gap_data[0]:
        position = gap_data[0].index (variable)
        for i in range(len(gap_data)-1):
            if variable == "country":
                return_list.append(gap_data[i+1][position])
            else: 
                return_list.append(float(gap_data[i+1][position]))
        return (return_list)
    else: 
        return (str(variable) +" is not a variable in this data. Please choose another variable")

# sample call command: 
select (gap_data, "gdpPercap")
    
```

```python
#bonus question
def select_lowincome(data, variable):
    '''
    The World Bank considers any country with GDP Per Capita lower than 1025 US dollars to be a low-income country.
    In using my select_lowincome function, when data and variable are entered properly, I try to identify low-income countries by their country names.
    So I added a boolean statement to append a True/False value when "gdpPercap" is selected.

    '''
    return_list = []
    if variable in gap_data[0]:
        position = gap_data[0].index (variable)
        for i in range(len(gap_data)-1):
            if variable == "country":
                return_list.append(gap_data[i+1][position])
            elif variable == "lifeExp":
                return_list.append(float(gap_data[i+1][position]))
            else: 
                return_list.append(float(gap_data[i+1][position])<1025)
        return (return_list)
    else: 
        return (str(variable) +" is not a variable in this data. Please choose another variable")

##sample call command:
mydic = dict(zip(select_lowincome(gap_data, "country"), select_lowincome(gap_data, "gdpPercap")))
mydic
    
```
