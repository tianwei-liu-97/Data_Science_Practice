```python
import pandas as pd
url = "https://raw.githubusercontent.com/edunford/coding_discussions_ppol564_fall2019/master/03_coding_discussion/chicago_summer_2018_crime_data.csv"
data = pd.read_csv (url)
```

```python
## first of all, I count the daily occurance of each crime 
month_day_data = \
(data
 .filter(['month','day','primary_type']) # select month, day, and primary_type
 .groupby(['month','day','primary_type']) 
 .size() # count the number of observations 
 .reset_index() # Reset to 0-based index
 .rename(columns={0:"month_day_count"}) # rename the column name of count
) 
month_day_data.head(10)
```

```python
# Count the monthly occurance of each crime, this will be merged with the previous 
# month_day_data to calculate the proportion of each crime each day of the month
m_data = \
(data
 .filter(['month','primary_type']) # Select month and primary_type
 .groupby(['month','primary_type'])
 .size() # count the number of observations each month
 .reset_index() #reset to 0_based index
 .rename(columns={0:"month_count"}) # rename the column name to month_count
) 
m_data.head(10)
```

```python
count_data = month_day_data.merge(m_data, how = "left") # Merge the two dataframes created above
                                                        # The reason that I choose "left merge" is because
                                                        # left merge will retian the the month_count within each month
count_data.head(10)
```

```python
count_data['proportion'] = (count_data.month_day_count / count_data.month_count).round(2)
        # Create a new column called "proportion" and round it to 2 decimal place
count_data.head(10)
```

```python
#Create a pivot table with primary_type and month as the index and day as column. Fill the table with proportion data
final_table = count_data.pivot_table('proportion',index = ['primary_type','month'], columns = 'day').fillna(0)
```

```python
data01 = data.filter(["month", "day", "primary_type"])

#count the times of every type of crime each day
data01.loc[:,"prime_count"] = 1
data02 = data01.groupby(["primary_type","month","day"]).count()

#reshape the data; use day as columns
data03 = data02.unstack(level="day")
data03 = data03.fillna(0)

#calulate the sum of every type of crime and save it to a new column
data03.loc[:,"prime_count_sum"] = data03.apply(lambda x: x.sum(),axis = 1)

#calculate the everyday's proportion of times
for i in range(31):
    data03.iloc[:,i] = data03.iloc[:,i]/data03.loc[:,"prime_count_sum"] 
    
primary_proportion = round(data03,2)
primary_proportion = primary_proportion.drop(columns = ['prime_count_sum'])
primary_proportion.head(20)
```
